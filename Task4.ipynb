```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

styles = [
    'default',
    'ggplot',
    'seaborn-v0_8-darkgrid',
    'bmh',
    'classic',
    'fivethirtyeight',
    'grayscale',
    'Solarize_Light2',
    'fast',
    'tableau-colorblind10'
]

for style in styles:
    plt.style.use(style)
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, label='sin(x)')
    plt.title(f"Matplotlib Style: {style}")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
