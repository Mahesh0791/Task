```python
import matplotlib.pyplot as plt
import pandas as pd

years = ['2004', '2006', '2008', '2010', '2012']
sachin_runs = [1000, 1200, 1100, 1300, 900]
kohli_runs = [500, 800, 1000, 1100, 1200]
sehwag_runs = [900, 950, 1050, 1200, 800]

df = pd.DataFrame({
    'Year': years,
    'Sachin': sachin_runs,
    'Kohli': kohli_runs,
    'Sehwag': sehwag_runs
})

bar_width = 0.25
x = range(len(df))

plt.figure(figsize=(10, 6))
plt.bar([i - bar_width for i in x], df['Sachin'], width=bar_width, label='Sachin')
plt.bar(x, df['Kohli'], width=bar_width, label='Kohli')
plt.bar([i + bar_width for i in x], df['Sehwag'], width=bar_width, label='Sehwag')
plt.xticks(x, df['Year'])
plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison Over Selected Years')
plt.legend()
plt.tight_layout()
plt.show()

players = ['Sachin', 'Kohli', 'Sehwag']
debut_runs = [4500, 4300, 4000]  

plt.figure(figsize=(8, 5))
plt.barh(players, debut_runs, color=['blue', 'green', 'orange'])
plt.xlabel('Runs')
plt.title('Total Runs in Debut 5 Years')
plt.tight_layout()
plt.show()
```

```python

labels = ['Android', 'iOS', 'Others']
market_share = [72, 26, 2]

plt.figure(figsize=(6, 6))
plt.pie(market_share, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Mobile OS Market Share')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
plt.barh(labels, market_share, color=['green', 'gray', 'red'])
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share (Bar View)')
plt.tight_layout()
plt.show()
```
